```python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm, trange


pd.options.display.max_columns = None
```

```python
dataset_raw = pd.read_excel("data.xlsx", sheet_name="spot_daily") \
    .melt(id_vars=["date"], var_name="help",value_name="price") \
    .assign(ticker=lambda x: x["help"].apply(lambda y: y.split("_")[0]),
            concept=lambda x: x["help"].apply(lambda y: y.split("_")[1])) \
    .pivot_table(index=["ticker","date"], columns=["concept"], values="price") \
    .rename_axis(None, axis=1) \
    .sort_index()

dataset_raw
```
